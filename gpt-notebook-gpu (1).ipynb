```python
train_df['original'] = train_df['Clinician'].copy()

def ques_index(x):
    index = x['Question_index']
    #add_string = ' QUESTIONS: what is the medical summary. '
    add_string = ' **QUESTIONS:** '
    prompt = x['Prompt']

    new = prompt[:index] + add_string + re.sub(r'^\s*questions?\s*\:?', ' ', prompt[index:], flags=re.IGNORECASE).strip()
    return re.sub(r'\s+', ' ', new).strip()



train_df['Prompt'] = train_df.apply(ques_index, axis=1)
# val_df['Prompt'] = val_df.apply(ques_index, axis=1)
test_df['Prompt'] = test_df.apply(ques_index, axis=1)


# ---------------------------------------------------------------------------------------------------

def res_index(x):
    index = x['Response_index']
    
    add_string = ' **ANSWERS:** '
    
    res = x['Clinician']

    new = res[:index] + add_string + res[index:]
    return re.sub(r'\s+', ' ', new).strip()

def res_index2(x):
    index = x['Response_index']
    
    # add_string = '</think><answer>'
    add_string = '</information><response>'
    
    res = x['original']

    new = res[:index] + add_string + res[index:] + "</response>"
    return re.sub(r'\s+', ' ', new).strip()




train_df['Clinician'] = train_df.apply(res_index, axis=1)
train_df['original'] = train_df.apply(res_index2, axis=1)




# ---------------------------------------------------------------------------------------------------




# summary
def summ(x):
    # remove the words summary.?, case when they appear at the begining of a string
    answer = re.sub(r'^case\s+', '', x, flags=re.IGNORECASE).lstrip()
    answer = re.sub(r'^summary\s*\:?\s*', '', answer, flags=re.IGNORECASE).lstrip()
    answer = re.sub(r'^summary\s*\.?\s*', '', answer, flags=re.IGNORECASE).lstrip()
    answer = re.sub(r'\s+', ' ', answer).strip()


    return '**SUMMARY:** ' + answer

def summ2(x):
    # remove the words summary.?, case when they appear at the begining of a string
    answer = re.sub(r'^case\s+', '', x, flags=re.IGNORECASE).lstrip()
    answer = re.sub(r'^summary\s*\:?\s*', '', answer, flags=re.IGNORECASE).lstrip()
    answer = re.sub(r'^summary\s*\.?\s*', '', answer, flags=re.IGNORECASE).lstrip()
    answer = re.sub(r'\s+', ' ', answer).strip()


    # return '<think>' + answer
    return '<information>' + answer
    # return answer



train_df['Clinician'] = train_df['Clinician'].apply(summ)
train_df['original'] = train_df['original'].apply(summ2)



# ---------------------------------------------------------------------------------------------------



# introduction

def insert_scene(x):
    x = kenya_pattern.sub(r'\1 ' + ' CLINICAL SCENARIO: ', x, count=1).strip()
    x = re.sub('\s+', ' ', x).strip()
    return x

train_df['Prompt'] = 'INTRODUCTION: ' + train_df['Prompt']
# val_df['Prompt'] = 'INTRODUCTION: ' + val_df['Prompt']
test_df['Prompt'] = 'INTRODUCTION: ' + test_df['Prompt']


# scenerio

train_df['Prompt'] = train_df['Prompt'].apply(insert_scene)
# val_df['Prompt'] = val_df['Prompt'].apply(insert_scene)
test_df['Prompt'] = test_df['Prompt'].apply(insert_scene)
```

```python
def replace_punct_with_space(text: str) -> str:
    """
    Replace all punctuation characters in `text` with a single space,
    collapse consecutive whitespace into a single space, and strip ends.

    Uses Unicode character categories so it handles ASCII and non-ASCII punctuation.
    """
    if not text:
        return text

    # Step 1 — replace punctuation characters (Unicode category 'P*') with a space
    chars = []
    for ch in text:
        if unicodedata.category(ch).startswith('P'):
            chars.append(' ')
        else:
            chars.append(ch)
    replaced = ''.join(chars)

    # Step 2 — collapse any sequence of whitespace into a single space and trim ends
    cleaned = re.sub(r'\s+', ' ', replaced).strip()
    return cleaned
```

# Embed Prompts for Retrieval

```python
from sentence_transformers import SentenceTransformer
import numpy as np
#import faiss
import hnswlib

# embedder = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2') # embedding_model_id = "NovaSearch/stella_en_400M_v5"
# train_embeddings = embedder.encode(train_df['Prompt'].tolist(), convert_to_numpy=True)

# # Create FAISS indexs
# dimension = train_embeddings.shape[1]
# index = faiss.IndexFlatL2(dimension)
# index.add(train_embeddings)

# Embedding and indexing (hnswlib)




# 1. Generate sentence embeddings
embedder = SentenceTransformer('sentence-transformers/all-roberta-large-v1') # try sentence-transformers/all-mpnet-base-v2 embedding_model_id = "sentence-transformers/all-roberta-large-v1", 'sentence-transformers/gtr-t5-xl'


train_texts = train_df['Prompt'].tolist()

train_embeddings = embedder.encode(train_texts, convert_to_numpy=True)

# 2. Initialize hnswlib index for cosine similarity (best for sentence embeddings)
dimension = train_embeddings.shape[1]
num_elements = len(train_embeddings)

index = hnswlib.Index(space='cosine', dim=dimension)  # 'cosine' is better for high-dimensional text embeddings

# 3. Maximize indexing quality
index.init_index(
    max_elements=num_elements,
    ef_construction=1000,  # extremely high for better accuracy (default ~200)
    M=100                 # higher connectivity = better accuracy (default 16)
)

# Add data to index
index.add_items(train_embeddings, ids=list(range(num_elements)))

# 4. Maximize search recall
index.set_ef(10000)  # ef >= 2 * top_k; higher = better recall (default ~10-100)
```

# Retrieval Function

```python
# Define retrieval function hnswlib
def retrieve_similar_context(df, top_k=3):
    query = df['Prompt']
    try:
        answer = df['Clinician']
    except:
        top_k = top_k-1

        query_vec = embedder.encode([query], convert_to_numpy=True)
        indices, distances = index.knn_query(query_vec, k=top_k)
        contexts = train_df.iloc[indices[0]]['Clinician'].tolist()
        questions = train_df.iloc[indices[0]]['Prompt'].tolist()
        
        new = list(zip(questions, contexts))
        
        
        #shuffle list to introduce randomness in order of apperance
        # random.shuffle(contexts)
        
        return new
        
    else:
        check = set()
        check2 =set()
        check.add(answer)
        check2.add(query)
        
        
        query_vec = embedder.encode([query], convert_to_numpy=True)
        indices, distances = index.knn_query(query_vec, k=top_k)
        contexts = set(train_df.iloc[indices[0]]['Clinician'].tolist())
        questions = set(train_df.iloc[indices[0]]['Prompt'].tolist())

        contexts = contexts - check
        questions = questions - check2

        contexts = list(contexts)
        questions = list(questions)
        
        new = list(zip(questions, contexts))
        
        # shuffle list to introduce randomness in order of apperance
        # random.shuffle(contexts)

        return new
```

```python
train_df["Examples"] = train_df.apply(retrieve_similar_context, axis=1)
test_df["Examples"] = test_df.apply(retrieve_similar_context, axis=1)
```
